```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Few-shot classification report data for first 10 classes
data = {
    'precision': [0.6667, 0.3671, 0.2118, 0.8043, 0.8, 0.9773, 0.9808, 0.5714, 1.0, 0.6173],
    'recall':    [0.4583, 0.7632, 0.3673, 0.8043, 0.3333, 0.8113, 1.0, 0.9362, 0.8636, 0.9434],
    'f1-score':  [0.5432, 0.4957, 0.2687, 0.8043, 0.4706, 0.8866, 0.9903, 0.7097, 0.9268, 0.7463],
    'support':   [48, 38, 49, 46, 36, 53, 51, 47, 44, 53]
}

classes = [f'Class {i}' for i in range(1, 11)]
df = pd.DataFrame(data, index=classes)

# Overall metrics
overall_metrics = {
    'accuracy': 0.6715,
    'macro avg precision': 0.6828,
    'macro avg recall': 0.6625,
    'macro avg f1-score': 0.6504,
    'weighted avg precision': 0.6898,
    'weighted avg recall': 0.6715,
    'weighted avg f1-score': 0.6594
}

# ---------------- Heatmap ----------------
plt.figure(figsize=(12,6))
sns.heatmap(df[['precision', 'recall', 'f1-score']], annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Few-Shot Classification Report (First 10 Classes)')
plt.ylabel('Classes')

# Add overall metrics as text below the heatmap
metrics_text = (f"Accuracy: {overall_metrics['accuracy']}\n"
                f"Macro Avg - Precision: {overall_metrics['macro avg precision']}, "
                f"Recall: {overall_metrics['macro avg recall']}, "
                f"F1-score: {overall_metrics['macro avg f1-score']}\n"
                f"Weighted Avg - Precision: {overall_metrics['weighted avg precision']}, "
                f"Recall: {overall_metrics['weighted avg recall']}, "
                f"F1-score: {overall_metrics['weighted avg f1-score']}")
plt.gcf().text(0.5, -0.1, metrics_text, ha='center', fontsize=10)

plt.show()
```
