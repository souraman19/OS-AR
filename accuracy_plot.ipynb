```python
import matplotlib.pyplot as plt

# Epochs (from 0 to 49 with given intervals)
epochs = [0, 4, 9, 14, 19, 24, 29, 34, 39, 44, 49]

# Corresponding test accuracies (in %)
accuracy = [1.8, 7.2, 16.8, 52.4, 67.4, 70.4, 71.6, 73.4, 74.6, 75.2, 75.2]

# Plot
plt.figure(figsize=(8, 5))
plt.plot(epochs, accuracy, marker='o', color='royalblue', linewidth=2, markersize=8)

# Labels and title
plt.title("X-CLIP One-Shot Training Accuracy over Epochs", fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)

# Optional enhancements
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(epochs)
plt.ylim(0, 100)

# Show the final plot
plt.tight_layout()
plt.show()


print("final test accuracy after one shot training : 75.0%")
```
